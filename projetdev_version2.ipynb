```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler


# Données d'entraînement (réduites aux 3 premières questions)
X = [
    [3, 0, 2],  # Exemple pour "Software Development"
    [1, 3, 3],  # Exemple pour "Data Science"
    [0, 2, 1],  # Exemple pour "System Administration"
    [2, 1, 0],  # Exemple pour "Cybersecurity"
]

# Correspondance des domaines d'intérêt
y = [0, 1, 2, 3]  # 0 = Software Development, 1 = Data Science, 2 = System Administration, 3 = Cybersecurity

# Normalisation des données d'entrée
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Créer un modèle k-NN avec plus de voisins
knn = KNeighborsClassifier(n_neighbors=3)

# Entraîner le modèle sur les données normalisées
knn.fit(X_scaled, y)

# Liste des domaines d'intérêt
domains = ["Software Development", "Data Science", "System Administration", "Cybersecurity"]

# Nouvelle réponse utilisateur (3 premières questions uniquement)
new_responses = [3, 3, 1]  # Exemple de réponses utilisateur

# Normaliser la réponse de l'utilisateur
new_responses_scaled = scaler.transform([new_responses])

# Prédire le domaine d'intérêt basé sur les réponses
predicted_domain_index = knn.predict(new_responses_scaled)[0]
predicted_domain = domains[predicted_domain_index]

# Afficher le domaine d'intérêt prédit
print(f"Domaine d'intérêt prédit : {predicted_domain}")

# Proposer des formations et certifications en fonction du domaine d'intérêt prédit
formations = {
    "Software Development": ["Formation en développement web", "Certifications Java, C#", "Formation en DevOps"],
    "Data Science": ["Formation en machine learning", "Certifications Python, R", "Formation en Big Data"],
    "System Administration": ["Formation en administration Linux", "Certifications Windows Server", "Formation en virtualisation"],
    "Cybersecurity": ["Formation en sécurité des réseaux", "Certifications CEH, CISSP", "Formation en cryptographie"]
}

# Afficher les formations et certifications suggérées
suggested_courses = formations[predicted_domain]
print("\nFormations et certifications suggérées :")
for course in suggested_courses:
    print(f"- {course}")
```
